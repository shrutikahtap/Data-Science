# Subplots
### You can create multiple plots in a single figure using subplots.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Creating subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# First subplot
ax1.plot(x, y1, 'r-o')
ax1.set_title('First Subplot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Second subplot
ax2.plot(x, y2, 'g-^')
ax2.set_title('Second Subplot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

# Adjusting layout to prevent overlap
plt.tight_layout()

# Displaying the plots
plt.show()
```

# Interactive Plots with 'mpl_toolkits'
### Matplotlib supports interactive plots, which can be useful for exploratory data analysis.

## 3D Plotting
## 3D plots can be created using 'mpl_toolkits.mplot3d'.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Creating a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Adding title and labels
ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Displaying the plot
plt.show()
```

# Advanced Techniques
### Twin Axes
### You can create plots with two different y-axes.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [1, 4, 9, 16, 25]

# Creating a plot with twin axes
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

# Adding labels
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y1-axis', color='g')
ax2.set_ylabel('Y2-axis', color='b')

# Displaying the plot
plt.show()
```

# Interactive Widgets with 'ipywidgets'
### For Jupyter Notebook users, 'ipywidgets' can add interactivity to Matplotlib plots.

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from ipywidgets import interact

# Function to plot a sine wave with variable frequency
def plot_sine_wave(frequency):
    x = np.linspace(0, 2 * np.pi, 1000)
    y = np.sin(frequency * x)
    plt.plot(x, y)
    plt.title('Interactive Sine Wave')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.ylim(-1.5, 1.5)
    plt.show()

# Creating an interactive widget
interact(plot_sine_wave, frequency=widgets.FloatSlider(value=1, min=0.1, max=5.0, step=0.1))
```

# Stacked Bar Plot
### Shows multiple bar plots stacked on top of each other.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values1 = [3, 7, 2]
values2 = [5, 3, 8]

bar_width = 0.5
index = range(len(categories))

plt.bar(index, values1, bar_width, label='Group 1', color='b')
plt.bar(index, values2, bar_width, bottom=values1, label='Group 2', color='r')

plt.xticks(index, categories)
plt.title('Stacked Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = tips['total_bill'].quantile(0.25)
Q3 = tips['total_bill'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Define the lower and upper bound for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print()
print(upper_bound)
```

```python
# Remove outliers
# method-1
filter1 = tips['total_bill'] > lower_bound
filter2 = tips['total_bill'] < upper_bound
tips_filtered = tips[filter1 & filter2]
```

```python
# Plot the box plot without outliers
sns.boxplot(x='day', y='total_bill', data=tips_filtered)
plt.show()
```
